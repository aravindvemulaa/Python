```python
#2. Accessing list

mylist = ['Banana', 'apple', 'guva', 'Orange','Mango','Grapes']
print(mylist[1])
print(mylist[2:])
print(mylist[:2])
print(mylist[2:5])

if 'apple' in mylist:
    print('Yes, Apple In the List')
else:
    print('apple is not in the list')
```

```python
#3. Changing/Updating List Items:

thislist = ['Java', 'Python', 'Sql', 'K8s', 'Docker','jenkins', 'Pega']
thislist[1] = 'AWS devops'
thislist[2:3] = ['pinapple', 'watermelon']
thislist.pop()
print(thislist)

thislist.insert(4, 'lemon')
thislist.append('kiwi')
print(thislist)

first_list = ['Ravi', 'suresh', 'naresh', 'tharun', 'srikanth']
second_list = ['Samantha', 'Keerthisuresh', 'anushka', 'thamanna', 'Sreelela']
second_list.extend(first_list)
print(second_list)
```

```python
#4. Removing items in the list
Newlist =['Tamato', 'beetroot', 'carrot', 'raddish', 'brinjal']
Newlist.pop() #brinjal
Newlist.remove('raddish') #raddish
Newlist.pop(2)   #carrot
del Newlist[1]   #beetroot
print(Newlist)
Newlist.clear()   # clears all the items in the list
print(Newlist)
```

```python
#5. Loops in Lists:

Constructor_list = ['here', 'we', 'have', 'double', 'circle', 'bracket', 'so', 'this', 'is', 'list']

#for loop
for i in range(len(Constructor_list)):
    print(Constructor_list[i])

#while loop
current_list = ['alpha', 'bravo','charlie', 'delta']
i = 0
while  i < len(current_list):
    print(current_list[i])
    i += 1
```
